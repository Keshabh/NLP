```python
paragraph = '''Narendra Damodardas Modi was born on 17 September 1950 to a Gujarati family of Other Backward Class (OBC) background and Hindu faith[13][14] in Vadnagar, Mehsana district, Bombay State (present-day Gujarat). He was the third of six children born to Damodardas Mulchand Modi (c. 1915–1989) and Hiraben Modi (1923–2022).[15][h][16] According to Modi and his neighbours, he worked infrequently in his father's tea stall in the Vadnagar railway station.[17][18][19]

Modi completed his higher secondary education in Vadnagar in 1967; his teachers described him as an average student and a keen, gifted debater with an interest in theatre.[20] He preferred playing larger-than-life characters in theatrical productions, which has influenced his political image.[21][22]
'''
```

```python
paragraph
```

```python
import nltk
from nltk.stem import PorterStemmer
from nltk.corpus import stopwords
nltk.download('stopwords')
```

```python
#tokenization- convert paragraph -> sentences -> words
nltk.download('punkt_tab')
```

```python
sentences = nltk.sent_tokenize(paragraph) #converts paragraph into sentences
sentences
```

```python
#stemming
stemmer = PorterStemmer()
stemmer.stem('hisotry')
```

```python
#lemmatization
from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')
lemmatizer = WordNetLemmatizer()
lemmatizer.lemmatize('history')
```

```python
#lets clean the data, and remove everything except a-ZA-Z
#lower the case
import re
corpus = []
for sentence in sentences:
  sentence = re.sub('[^a-zA-Z]',' ',sentence)
  corpus.append(sentence.lower())
corpus
```

```python
stopwords.words('english')
```

```python
len(corpus)
```

```python
#Apply stemming
corpus_final = []
for i in corpus:
  #convert sentences into words
  words = nltk.word_tokenize(i)
  temp = []
  for word in words:
    if word not in set(stopwords.words('english')):
      # print(stemmer.stem(word))
      temp.append(lemmatizer.lemmatize(word))
  corpus_final.append(' '.join(temp))
corpus_final
```

```python
#bag of words
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(binary=True)
x = cv.fit_transform(corpus_final)
```

```python
cv.vocabulary_
```

```python
#this is for first sentence in corpus
print(corpus[0])
x[0].toarray()
```

```python
#tf-idf
from sklearn.feature_extraction.text import TfidfVectorizer
tf = TfidfVectorizer(ngram_range=(1,3), max_features =30)
#max features gives top 30 highest frequency words in the features
y = tf.fit_transform(corpus_final)
y
```
